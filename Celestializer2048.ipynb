```python
from tokenizers import Tokenizer, decoders, models, normalizers, pre_tokenizers, trainers
from datasets import load_dataset
dataset = load_dataset("Open-Orca/OpenOrca")["train"]
```

```python
tokenizer = Tokenizer(models.BPE())
tokenizer.normalizer = normalizers.NFKC()
tokenizer.pre_tokenizer = pre_tokenizers.ByteLevel()
tokenizer.decoder = decoders.ByteLevel()
trainer = trainers.BpeTrainer(
    vocab_size=2048,
    initial_alphabet=pre_tokenizers.ByteLevel.alphabet(),
    special_tokens=[ "<[PAD]>","<[USER]>", "<[MODEL]>"],
    show_progress=True
)
```

```python
def ds_batch_iterator(batch_size):
    for i in range(0, len(dataset), batch_size):
        yield dataset[i:i+batch_size]["question"]
```

```python
tokenizer.train_from_iterator(ds_batch_iterator(1000), trainer=trainer, length=len(dataset))
tokenizer.save("Celestializer2048.json")
```

```python
r = tokenizer.encode("My name is Runna ! Nice to meet you !")
print(r.ids)
print(r.tokens)
```
